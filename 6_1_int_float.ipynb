```python
""" –ü–ª–æ—â–∞–¥—å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç –¥–ª–∏–Ω—ã –¥–≤—É—Ö –∫–∞—Ç–µ—Ç–æ–≤ –≤ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–æ–º —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–µ –∏ –≤—ã–≤–æ–¥–∏—Ç –µ–≥–æ –ø–ª–æ—â–∞–¥—å. 
–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –¥–≤–∞ —á–∏—Å–ª–∞ —Å –ø–ª–∞–≤–∞—é—â–µ–π —Ç–æ—á–∫–æ–π ‚Äì –¥–ª–∏–Ω—ã –∫–∞—Ç–µ—Ç–æ–≤, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ. """ 
a, b, = float(input()), float(input())
s = 0.5 * a * b
print(s)
```

```python
"""–î–≤–∞ –∞–≤—Ç–æ–º–æ–±–∏–ª—è 
–î–≤–∞ –∞–≤—Ç–æ–º–æ–±–∏–ª—è –µ–¥—É—Ç –Ω–∞–≤—Å—Ç—Ä–µ—á—É –¥—Ä—É–≥ –¥—Ä—É–≥—É —Å –ø–æ—Å—Ç–æ—è–Ω–Ω—ã–º–∏ —Å–∫–æ—Ä–æ—Å—Ç—è–º–∏ V_1, V_2 –∫–º/—á. –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —á–µ—Ä–µ–∑ –∫–∞–∫–æ–µ –≤—Ä–µ–º—è –∞–≤—Ç–æ–º–æ–±–∏–ª–∏ –≤—Å—Ç—Ä–µ—Ç—è—Ç—Å—è, –µ—Å–ª–∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –Ω–∏–º–∏ —Ä–∞–≤–Ω–æ S –∫–º.
–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —Ç—Ä–∏ —á–∏—Å–ª–∞ —Å –ø–ª–∞–≤–∞—é—â–µ–π —Ç–æ—á–∫–æ–π S, V1, V2, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ."""
s, b, a = float(input()), float(input()), float(input())
print(s / (a + b))
```

```python
"""–û–±—Ä–∞—Ç–Ω–æ–µ —á–∏—Å–ª–æ
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã –æ–¥–Ω–æ —á–∏—Å–ª–æ –∏ –≤—ã–≤–æ–¥–∏—Ç –æ–±—Ä–∞—Ç–Ω–æ–µ –µ–º—É. –ï—Å–ª–∏ –ø—Ä–∏ —ç—Ç–æ–º –≤–≤–µ–¥—ë–Ω–Ω–æ–µ —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã —á–∏—Å–ª–æ ‚Äì –Ω–æ–ª—å, —Ç–æ –≤—ã–≤–µ—Å—Ç–∏ ¬´–û–±—Ä–∞—Ç–Ω–æ–≥–æ —á–∏—Å–ª–∞ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç¬ª (–±–µ–∑ –∫–∞–≤—ã—á–µ–∫).
–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ."""
a = float(input())
if a == 0:
    print('–û–±—Ä–∞—Ç–Ω–æ–≥–æ —á–∏—Å–ª–∞ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç')
else:
    print(a**(-1))
```

```python
"""451 –≥—Ä–∞–¥—É—Å –ø–æ –§–∞—Ä–µ–Ω–≥–µ–π—Ç—É 
–£ –∏–∑–≤–µ—Å—Ç–Ω–æ–≥–æ –∞–º–µ—Ä–∏–∫–∞–Ω—Å–∫–æ–≥–æ –ø–∏—Å–∞—Ç–µ–ª—è –†—ç—è –ë—Ä–µ–¥–±–µ—Ä–∏ –µ—Å—Ç—å —Ä–æ–º–∞–Ω ¬´451 –≥—Ä–∞–¥—É—Å –ø–æ –§–∞—Ä–∞–Ω–≥–µ–π—Ç—É¬ª. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –∫–∞–∫–æ–π —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–µ –ø–æ —à–∫–∞–ª–µ –¶–µ–ª—å—Å–∏—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —à–∫–∞–ª–µ –§–∞—Ä–µ–Ω–≥–µ–π—Ç–∞.

–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ–æ—Ä–º—É–ª—É –¥–ª—è –ø–µ—Ä–µ–≤–æ–¥–∞: C = 5/9 * (F‚àí 32).

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ –≥—Ä–∞–¥—É—Å–æ–≤ –ø–æ —à–∫–∞–ª–µ –§–∞—Ä–µ–Ω–≥–µ–π—Ç–∞."""

f = float(input())
print((5 / 9) * (f-32))
```

```python
"""Dog age
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —á–∏—Å–ª–æ nn ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±–∞—á—å–∏—Ö –ª–µ—Ç. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –≤–æ–∑—Ä–∞—Å—Ç —Å–æ–±–∞–∫–∏ –≤ —á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏—Ö –≥–æ–¥–∞—Ö.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±–∞—á—å–∏—Ö –ª–µ—Ç.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –≤–æ–∑—Ä–∞—Å—Ç —Å–æ–±–∞–∫–∏ –≤ —á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏—Ö –≥–æ–¥–∞—Ö.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –í —Ç–µ—á–µ–Ω–∏–µ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –ª–µ—Ç —Å–æ–±–∞—á–∏–π –≥–æ–¥ —Ä–∞–≤–µ–Ω 10.510.5 —á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏–º –≥–æ–¥–∞–º. –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –∫–∞–∂–¥—ã–π –≥–æ–¥ —Å–æ–±–∞–∫–∏ —Ä–∞–≤–µ–Ω 4 —á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏–º –≥–æ–¥–∞–º."""

f = float(input())
if f <=2:
    print(f * 10.5)
else:
    print((2 * 10.5)+(f-2)*4)
```

```python
"""–ü–µ—Ä–≤–∞—è —Ü–∏—Ñ—Ä–∞ –ø–æ—Å–ª–µ —Ç–æ—á–∫–∏
–î–∞–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –í—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ –ø–µ—Ä–≤—É—é —Ü–∏—Ñ—Ä—É –ø–æ—Å–ª–µ –¥–µ—Å—è—Ç–∏—á–Ω–æ–π —Ç–æ—á–∫–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ü–∏—Ñ—Ä—É –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏."""

a = float(input())
a = int((a * 10) % 10)
print(a)
```

```python
"""–î—Ä–æ–±–Ω–∞—è —á–∞—Å—Ç—å
–î–∞–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –í—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ –¥—Ä–æ–±–Ω—É—é —á–∞—Å—Ç—å.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ."""
a = float(input())
print(a % 1)
```

```python
"""–ù–∞–∏–±–æ–ª—å—à–µ–µ –∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∏–∑ –ø—è—Ç–∏ —á–∏—Å–µ–ª.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –ø—è—Ç—å —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ."""

a, b , c, d, e = int(input()), int(input()), int(input()), int(input()), int(input())
print('–ù–∞–∏–º–µ–Ω—å—à–µ–µ —á–∏—Å–ª–æ =', min(a, b, c, d, e))
print('–ù–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ =', max(a, b, c, d, e))
```

```python
"""–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ —Ç—Ä—ë—Ö üå∂Ô∏è
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–µ—Ç —Ç—Ä–∏ —á–∏—Å–ª–∞ –æ—Ç –±–æ–ª—å—à–µ–≥–æ –∫ –º–µ–Ω—å—à–µ–º—É."""
a, b , c = int(input()), int(input()), int(input())
print(max(a, b , c))
print(a + b + c - max(a, b , c) - min(a, b , c))
print(min(a, b , c))
```
