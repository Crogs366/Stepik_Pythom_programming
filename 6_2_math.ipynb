```python
"""–ï–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ
–ù–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏ (x1: y1) –∏ (x2 ; y2) )–æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Ç–∞–∫ p = sqrt{(x_1-x_2)^2 + (y_1-y_2)^2})
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏, –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∫–æ—Ç–æ—Ä—ã—Ö –∑–∞–¥–∞–Ω—ã.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —á–µ—Ç—ã—Ä–µ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äì x1, y1, x2, y2. """
from math import *

x1, y1, x2, y2 = float(input()), float(input()), float(input()), float(input())
p = sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
print(p)
```

```python
"""–ü–ª–æ—â–∞–¥—å –∏ –¥–ª–∏–Ω–∞
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é –ø–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞ –∏ –¥–ª–∏–Ω—É –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É —Ä–∞–¥–∏—É—Å—É R.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ R."""

from math import *

r = float(input())
print(pi * r ** 2) #  S
print(2 * pi * r)
```

```python
"""–°—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è
–í –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ –≤—ã–¥–µ–ª—è—é—Ç —Å–ª–µ–¥—É—é—â–∏–µ —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è:

—Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —á–∏—Å–µ–ª a –∏ b
—Å—Ä–µ–¥–Ω–µ–µ –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ —á–∏—Å–µ–ª aa –∏ bb
—Å—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ —á–∏—Å–µ–ª aa –∏ bb
—Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ —á–∏—Å–µ–ª aa –∏ bb
–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ aa –∏ bb‚Äã, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ."""

from math import *

a, b = float(input()), float(input())
print((a + b) / 2)
print((a * b) ** 0.5)
print(2 * a * b / (a + b))
print(sqrt((a ** 2 + b ** 2) / 2))
```

```python
"""–¢—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã—á–∏—Å–ª—è—é—â—É—é –∑–Ω–∞—á–µ–Ω–∏–µ —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è
sin x + cos x + tan^2 x
–ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É —á–∏—Å–ª—É –≥—Ä–∞–¥—É—Å–æ–≤ x.
–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ xx –∏–∑–º–µ—Ä—è–µ–º–æ–µ –≤ –≥—Ä–∞–¥—É—Å–∞—Ö‚Äã. 
–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ
–¢—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∏–Ω–∏–º–∞—é—Ç –∞—Ä–≥—É–º–µ–Ω—Ç –≤ —Ä–∞–¥–∏–∞–Ω–∞—Ö. –ß—Ç–æ–±—ã –ø–µ—Ä–µ–≤–µ—Å—Ç–∏ –≥—Ä–∞–¥—É—Å—ã –≤ —Ä–∞–¥–∏–∞–Ω—ã, –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å —Ñ–æ—Ä–º—É–ª–æ–π 
r = (x * p) / 180‚Äã	"""

from math import *

x = float(input())
r = (x * pi) / 180 
print(sin(r) + cos(r) + tan(r) ** 2)
```

```python
"""–ü–æ–ª –∏ –ø–æ—Ç–æ–ª–æ–∫
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã—á–∏—Å–ª—è—é—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ ‚åàx‚åâ +‚åäx‚åã –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–º—É —á–∏—Å–ª—É x.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ xx.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. ‚åàx‚åâ ‚Äì –ø–æ—Ç–æ–ª–æ–∫ —á–∏—Å–ª–∞, [x‚åã ‚Äì –ø–æ–ª —á–∏—Å–ª–∞."""

from math import *

x = float(input())
print(floor(x) + ceil(x))
```

```python
"""–ö–≤–∞–¥—Ä–∞—Ç–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ üå∂Ô∏èüå∂Ô∏è
–î–∞–Ω—ã —Ç—Ä–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ a, b, c. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –∫–æ—Ä–Ω–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è 
ax^2 + bx + c = 0.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Ç—Ä–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ a,b,c, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ï—Å–ª–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏–º–µ–µ—Ç –¥–≤–∞ –∫–æ—Ä–Ω—è, —Ç–æ —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ –∏—Ö –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è."""

from math import *

a, b, c = float(input()), float(input()), float(input())
D = b ** 2 - 4 * a * c
if D > 0:
    x1 = (-b + sqrt(D)) / (2 * a)
    x2 = (-b - sqrt(D)) / (2 * a)
    print(min(x1, x2))
    print(max(x1, x2))
elif D == 0:
    print(-b / (2 * a))
else:
    print('–ù–µ—Ç –∫–æ—Ä–Ω–µ–π')
```

```python
"""–ü—Ä–∞–≤–∏–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫
–ü—Ä–∞–≤–∏–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫ ‚Äî –≤—ã–ø—É–∫–ª—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫, —É –∫–æ—Ç–æ—Ä–æ–≥–æ —Ä–∞–≤–Ω—ã –≤—Å–µ —Å—Ç–æ—Ä–æ–Ω—ã –∏ –≤—Å–µ —É–≥–ª—ã –º–µ–∂–¥—É —Å–º–µ–∂–Ω—ã–º–∏ —Å—Ç–æ—Ä–æ–Ω–∞–º–∏. –ü–ª–æ—â–∞–¥—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ —Å –¥–ª–∏–Ω–æ–π —Å—Ç–æ—Ä–æ–Ω—ã aa –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Å—Ç–æ—Ä–æ–Ω nn –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ: 
S = (n * a^2)/(4tg(Pi/n))

–î–∞–Ω—ã –¥–≤–∞ —á–∏—Å–ª–∞: –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ nn –∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ aa. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –ø–ª–æ—â–∞–¥—å —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ —á–∏—Å–ª–∞ nn –∏ aa, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ."""

from math import *

n, a = int(input()), float(input())
s = (n * a ** 2) / (4 * tan(pi / n))
print(s) 
```
